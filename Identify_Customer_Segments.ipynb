#### Step 1.1.1: Convert Missing Value Codes to NaNs
The fourth column of the feature attributes summary (loaded in above as `feat_info`) documents the codes from the data dictionary that indicate missing or unknown data. While the file encodes this as a list (e.g. `[-1,0]`), this will get read in as a string object. You'll need to do a little bit of parsing to make use of it to identify and clean the data. Convert data that matches a 'missing' or 'unknown' value code into a numpy NaN value. You might want to see how much data takes on a 'missing' or 'unknown' code, and how much data is naturally missing, as a point of interest.

**As one more reminder, you are encouraged to add additional cells to break up your analysis into manageable chunks.**

```python
# Identify missing or unknown data values and convert them to NaNs.
num_rows = df_azdias.shape[0]

naturally_missing = []
encoded_missing = []
total_missing = []

for attribute in feat_info['attribute']:
    # Count Data Naturally Missing
    num_nans = df_azdias[attribute].isnull().sum()

    # Count Data Encoded Missing
    feat_info_attribute = feat_info[feat_info['attribute'] == attribute]
    missing_value_codes = feat_info_attribute['missing_or_unknown'][feat_info_attribute.index[0]] # retrieve missing value codes
    if missing_value_codes == '[]':
        pass
    else:
        if 'X' in missing_value_codes:
            missing_value_codes = [val for val in missing_value_codes.strip('[]').split(',')] # convert string to list of strings
        else:
            missing_value_codes = [int(val) for val in missing_value_codes.strip('[]').split(',')] # convert string to list of integers
        num_encoded = df_azdias[attribute].isin(missing_value_codes).sum()

    # Convert Missing Value Codes to NaNs
    df_azdias[attribute] = df_azdias[attribute].replace(missing_value_codes, np.nan)

    # Count Total Data Missing
    num_total = df_azdias[attribute].isnull().sum()

    print(f"Attribute: {attribute}")
    if num_total == 0:
        naturally_missing.append(0)
        encoded_missing.append(0)
        total_missing.append(0)
        print("NA\n")
    else:
        naturally_missing.append(num_nans)
        print(f"Data Naturally Missing: {num_nans};\t\tPercentage: {num_nans/num_rows}")

        if missing_value_codes == '[]':
            encoded_missing.append(0)
            print(f"Data Encoded Missing: NA;\t\tPercentage: NA")
        else:
            encoded_missing.append(num_encoded)
            print(f"Data Encoded Missing: {num_encoded};\t\tPercentage: {num_encoded/num_rows}")
        
        total_missing.append(num_total)
        print(f"Total Data Missing: {num_total};\t\tPercentage: {num_total/num_rows}\n")

feat_info['naturally_missing'] = naturally_missing
feat_info['encoded_missing'] = encoded_missing
feat_info['total_missing'] = total_missing
feat_info['total_missing_perc'] = [val/num_rows for val in total_missing]
```

### Step 1.3: Create a Cleaning Function

Even though you've finished cleaning up the general population demographics data, it's important to look ahead to the future and realize that you'll need to perform the same cleaning steps on the customer demographics data. In this substep, complete the function below to execute the main feature selection, encoding, and re-engineering steps you performed above. Then, when it comes to looking at the customer data in Step 3, you can just run this function on that DataFrame to get the trimmed dataset in a single step.

```python
def clean_data(df, feat_info, categorical_features_encode, categorical_features_none, mixed_features_encode, other_features, decade_dict, movement_dict, wealth_dict, lifestage_dict):
    """
    Perform feature trimming, re-encoding, and engineering for demographics data
    
    INPUT: Demographics DataFrame
    OUTPUT: Trimmed and cleaned demographics DataFrame
    """
    
    # Put in code here to execute all main cleaning steps:
    # convert missing value codes into NaNs, ...
    for attribute in feat_info['attribute']:       
        feat_info_attribute = feat_info[feat_info['attribute'] == attribute]
        missing_value_codes = feat_info_attribute['missing_or_unknown'][feat_info_attribute.index[0]] # retrieve missing value codes
        if missing_value_codes == '[]':
            pass
        else:
            if 'X' in missing_value_codes:
                missing_value_codes = [val for val in missing_value_codes.strip('[]').split(',')] # convert string to list of strings
            else:
                missing_value_codes = [int(val) for val in missing_value_codes.strip('[]').split(',')] # convert string to list of integers

        # Convert Missing Value Codes to NaNs
        df[attribute] = df[attribute].replace(missing_value_codes, np.nan)

    # remove selected columns and rows, ...
    df = df.drop(labels=outlier_columns, axis=1)
    missing_per_row = df.isnull().sum(axis=1)
    few_missing_rows_index = missing_per_row[missing_per_row < 20].index
    df = df.iloc[few_missing_rows_index]
        
    # select, re-encode, and engineer column values.
    df_categorical_features_rencoded = df['OST_WEST_KZ'].replace({'O': 0, 'W': 1})
    
    encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
    df_categorical_features_encoded = pd.DataFrame(encoder.fit_transform(df[categorical_features_encode]))
    df_categorical_features_encoded.index = df.index
    df_categorical_features_encoded.columns = encoder.get_feature_names_out()

    df_mixed_features_split_encoded = pd.DataFrame()
    df_mixed_features_split_encoded['decade'] = df['PRAEGENDE_JUGENDJAHRE'].map(decade_dict)
    df_mixed_features_split_encoded['movement'] = df['PRAEGENDE_JUGENDJAHRE'].map(movement_dict)

    df_mixed_features_split = pd.DataFrame()
    df_mixed_features_split['wealth'] = df['CAMEO_INTL_2015'].map(wealth_dict)
    df_mixed_features_split['lifestage'] = df['CAMEO_INTL_2015'].map(lifestage_dict)

    df_mixed_features_encoded = pd.DataFrame(encoder.fit_transform(df[mixed_features_encode]))
    df_mixed_features_encoded.index = df.index
    df_mixed_features_encoded.columns = encoder.get_feature_names_out()

    # Return the cleaned dataframe.
    df = df[categorical_features_none + other_features]
    df = pd.concat([df, df_categorical_features_rencoded, df_categorical_features_encoded, df_mixed_features_split_encoded, df_mixed_features_split, df_mixed_features_encoded], axis=1)
    return df
```

### Step 2.3: Interpret Principal Components

Now that we have our transformed principal components, it's a nice idea to check out the weight of each variable on the first few components to see if they can be interpreted in some fashion.

As a reminder, each principal component is a unit vector that points in the direction of highest variance (after accounting for the variance captured by earlier principal components). The further a weight is from zero, the more the principal component is in the direction of the corresponding feature. If two features have large weights of the same sign (both positive or both negative), then increases in one tend expect to be associated with increases in the other. To contrast, features with different signs can be expected to show a negative correlation: increases in one variable should result in a decrease in the other.

- To investigate the features, you should map each weight to their corresponding feature name, then sort the features according to weight.
